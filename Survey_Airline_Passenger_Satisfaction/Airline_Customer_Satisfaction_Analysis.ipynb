```python
import numpy as np

# Convert columns to appropriate data types
airline_satisfaction_data['ID'] = airline_satisfaction_data['ID'].astype('category')
airline_satisfaction_data['Gender'] = airline_satisfaction_data['Gender'].astype('category')
airline_satisfaction_data['Age'] = airline_satisfaction_data['Age'].astype(int)
airline_satisfaction_data['Customer Type'] = airline_satisfaction_data['Customer Type'].astype('category')
airline_satisfaction_data['Type of Travel'] = airline_satisfaction_data['Type of Travel'].astype('category')
airline_satisfaction_data['Class'] = airline_satisfaction_data['Class'].astype('category')
airline_satisfaction_data['Flight Distance'] = airline_satisfaction_data['Flight Distance'].astype(int)
airline_satisfaction_data['Departure Delay'] = airline_satisfaction_data['Departure Delay'].astype(int)
airline_satisfaction_data['Arrival Delay'] = airline_satisfaction_data['Arrival Delay'].astype(float)

# Define a logical order for categorical variables
ordered_categories = [1, 2, 3, 4, 5]

# Convert specified columns to ordered categorical types
columns_to_order = [
    'Departure and Arrival Time Convenience', 'Ease of Online Booking', 'Check-in Service', 
    'Online Boarding', 'Gate Location', 'On-board Service', 'Seat Comfort', 'Leg Room Service', 
    'Cleanliness', 'Food and Drink', 'In-flight Service', 'In-flight Wifi Service', 
    'In-flight Entertainment', 'Baggage Handling'
]

# Replace 0 with NaN in specified columns
airline_satisfaction_data[columns_to_order] = airline_satisfaction_data[columns_to_order].replace(0, np.nan)

for column in columns_to_order:
    airline_satisfaction_data[column] = pd.Categorical(airline_satisfaction_data[column], categories=ordered_categories, ordered=True)

airline_satisfaction_data['Satisfaction'] = airline_satisfaction_data['Satisfaction'].astype('category')

# Display updated data types
print("\nUpdated Data Types:\n", airline_satisfaction_data.dtypes)
```

### Checking for outliers if any
Below is the updated markdown code to include checking for outliers using box plots:

```python
# Check for outliers using box plots
numeric_columns = airline_satisfaction_data.select_dtypes(include=[np.number]).columns.tolist()

# Set up the figure and axes for subplots
fig, axes = plt.subplots(nrows=(len(numeric_columns) + 1) // 2, ncols=2, figsize=(8, 2 * ((len(numeric_columns) + 1) // 2)))

# Flatten the axes array for easy iteration
axes = axes.flatten()

# Create box plots for each numeric column
for i, column in enumerate(numeric_columns):
    airline_satisfaction_data.boxplot(column=column, ax=axes[i])
    axes[i].set_title(f'Box Plot of {column}')

# Remove any empty subplots
for j in range(i + 1, len(axes)):
    fig.delaxes(axes[j])

plt.tight_layout()
plt.show()
```

```python
# Step 2: Statistical detection of outliers using IQR
outliers = {}

for feature in numeric_columns:
    Q1 = airline_satisfaction_data[feature].quantile(0.05)
    Q3 = airline_satisfaction_data[feature].quantile(0.95)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers[feature] = airline_satisfaction_data[(airline_satisfaction_data[feature] < lower_bound) | (airline_satisfaction_data[feature] > upper_bound)]

# Print the number of outliers in each feature
for feature, outlier_data in outliers.items():
    print(f'{feature}: {len(outlier_data)} outliers')
```

### Using only the numerical variables are not sufficient to differentiate the clusters. 

`KModes` is a clustering algorithm used in data science to group similar data points into clusters based on their categorical attributes. Unlike traditional clustering algorithms that use distance metrics, `KModes` works by identifying the modes or most frequent values within each cluster to determine its centroid. `KModes` is ideal for clustering categorical data such as customer demographics, market segments, or survey responses.

`KMeans` uses mathematical measures (distance) to cluster continuous data. The lesser the distance, the more similar our data points are.

# 4. Clustering Modeling
NEXT, we will use k-prototypes clustering from the <code>kmodes</code> library, which is designed to handle both numerical and categorical data.

<b>How Does the KModes Algorithm Work?</b>
<ol>
    <li>Pick K observations at random and use them as leaders/clusters</li>
    <li>Calculate the dissimilarities and assign each observation to its closest cluster</li>
    <li>Define new modes for the clusters</li>
    <li>Repeat 2–3 steps until there are is no re-assignment required</li>
</ol>

```python
from kmodes.kprototypes import KPrototypes

# Convert all columns to categorical for K-Prototypes
cat_cols = [df_clustering.columns.get_loc(col) for col in df_clustering.select_dtypes(['category', 'object']).columns]
for col in df_clustering.columns[cat_cols]:
    df_clustering[col] = df_clustering[col].astype('str')

# Apply K-Prototypes clustering
kproto = KPrototypes(n_clusters=4, init='Cao', verbose=2)
clusters = kproto.fit_predict(df_clustering, categorical=cat_cols)

# Add cluster labels to the dataframe
df_clustering['Cluster'] = clusters

# Print the cluster centroids
print("Cluster Centroids:")
print(kproto.cluster_centroids_)

# Visualize the clusters
sns.pairplot(df_clustering, hue='Cluster', palette='Set2', markers=["o", "s", "D"])
plt.show()
```

```python
from sklearn.metrics import silhouette_samples, silhouette_score
import numpy as np
from sklearn.preprocessing import LabelEncoder

# Convert all object and category data variables to numerical
label_encoders = {}
for column in df_clustering.select_dtypes(include=['category', 'object']).columns:
    le = LabelEncoder()
    df_clustering[column] = le.fit_transform(df_clustering[column])
    label_encoders[column] = le

def plot_silhouette(X, cluster_labels, title):
    n_clusters = len(set(cluster_labels))
    silhouette_avg = silhouette_score(X, cluster_labels)
    sample_silhouette_values = silhouette_samples(X, cluster_labels)

    y_lower = 10
    plt.figure(figsize=(10, 6))
    
    for i in range(n_clusters):
        ith_cluster_silhouette_values = sample_silhouette_values[cluster_labels == i]
        ith_cluster_silhouette_values.sort()

        size_cluster_i = ith_cluster_silhouette_values.shape[0]
        y_upper = y_lower + size_cluster_i

        plt.fill_betweenx(np.arange(y_lower, y_upper),
                          0, ith_cluster_silhouette_values,
                          alpha=0.7)

        plt.text(-0.05, y_lower + 0.5 * size_cluster_i, str(i))

        y_lower = y_upper + 10

    plt.title(title)
    plt.xlabel("Silhouette coefficient values")
    plt.ylabel("Cluster label")

    plt.axvline(x=silhouette_avg, color="red", linestyle="--")
    plt.yticks([])
    plt.show()

# Silhouette plot for clustering without PCA
plot_silhouette(df_clustering, df_clustering['Cluster'], 'Silhouette Plot for Clustering with KModes')
```

<b>Silhouette Analysis</b>
<ul>

# 4. Clustering Profiling
As mentioned in the previous section, this section will perform cluster profiling to identify the characteristics of the clusters that have been created by K-Means.

### K-means Clustering Results Summary
For illustration, we summarize the K-Means clustering results as below:

```python
# Function to create a summary table for clusters
def create_cluster_summary(df, cluster_column, features):
    summary_list = []

    for feature in features:
        for metric in ['mean', 'median', 'std']:
            row = {'Column Name': feature, 'Metric': metric}
            row['Overall'] = getattr(df[feature], metric)()
            for cluster in df[cluster_column].unique():
                if metric == 'mean':
                    row[f'Cluster {cluster+1}'] = df[df[cluster_column] == cluster][feature].mean()
                elif metric == 'median':
                    row[f'Cluster {cluster+1}'] = df[df[cluster_column] == cluster][feature].median()
                elif metric == 'std':
                    row[f'Cluster {cluster+1}'] = df[df[cluster_column] == cluster][feature].std()
            summary_list.append(row)
    
    summary_df = pd.DataFrame(summary_list)
    return summary_df

# Create a summary table for K-Means clusters
kmeans_summary_table = create_cluster_summary(df_clustering, 'Cluster', numeric_columns)

# Highlight the values in the summary table
highlighted_kmeans_summary_table = kmeans_summary_table.style.background_gradient(cmap='YlOrBr').set_properties(**{'text-align': 'right'}).hide(axis='index')

# Display the highlighted summary table
highlighted_kmeans_summary_table
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix, precision_score, recall_score, f1_score

# Initialize models
models = {
    'Logistic Regression': LogisticRegression(max_iter=1000),
    'Decision Tree': DecisionTreeClassifier(random_state=42),
    'Random Forest': RandomForestClassifier(random_state=42),
    'XGBoost': XGBClassifier(eval_metric='logloss', random_state=42)
}

# DataFrame to store results
results = pd.DataFrame(columns=['Model', 'Accuracy', 'Precision', 'Recall', 'F1-score'])

# Train and evaluate models
model_results = []

for model_name, model in models.items():
    print(f"Training {model_name}...")
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    
    accuracy = accuracy_score(y_test, y_pred)
    precision = precision_score(y_test, y_pred)
    recall = recall_score(y_test, y_pred)
    f1 = f1_score(y_test, y_pred)
    
    # Append the results to the list
    model_results.append({
        'Model': model_name,
        'Accuracy': accuracy,
        'Precision': precision,
        'Recall': recall,
        'F1-score': f1
    })

# Convert the list to a DataFrame
results = pd.DataFrame(model_results)

# Display the results
print(results)
```

### Model Optimization
We will GridSearchCV to optimize the models, try various hyper-parameters.

<b> Task 4: Model Tuning </b>

To optimize the models using <code>GridSearchCV</code>, you can specify a grid of hyperparameters for each model and use GridSearchCV to find the best combination of hyperparameters.
<hr>
<blockquote>
    <ul>
    <li><b>Define Parameter Grids</b>: Define the parameter grid for each model to be used in <code>GridSearchCV</code></li>
    <li><b>Optimize, and Evaluate Models</b>: A comprehensive evaluation of each model's performance based on the best found hyperparameters.</li>
    </ul>

```python
# Define the parameter grid for each model
param_grid_lr = {
    'C': [0.1, 1, 10, 100],
    'solver': ['liblinear', 'saga']
}

param_grid_dt = {
    'max_depth': [3, 5, 10, None],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]
}

param_grid_rf = {
    'n_estimators': [100, 200, 300],
    'max_depth': [3, 5, 10, None],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]
}

param_grid_xgb = {
    'n_estimators': [100, 200, 300],
    'max_depth': [3, 5, 10],
    'learning_rate': [0.01, 0.1, 0.2],
    'subsample': [0.6, 0.8, 1.0]
}
```

```python
from sklearn.model_selection import GridSearchCV

# Initialize models
models = {
    'Logistic Regression': (LogisticRegression(max_iter=1000), param_grid_lr),
    'Decision Tree': (DecisionTreeClassifier(random_state=42), param_grid_dt),
    'Random Forest': (RandomForestClassifier(random_state=42), param_grid_rf),
    'XGBoost': (XGBClassifier(eval_metric='logloss', random_state=42), param_grid_xgb)
}

# DataFrame to store results
results = pd.DataFrame(columns=['Model', 'Best Params', 'Accuracy', 'Precision', 'Recall', 'F1-score'])

# Train, optimize and evaluate models
model_results = []
feature_importance = {}

for model_name, (model, param_grid) in models.items():
    print(f"Optimizing {model_name}...")
    grid_search = GridSearchCV(estimator=model, param_grid=param_grid, cv=5, scoring='f1', n_jobs=-1)
    grid_search.fit(X_train, y_train)
    best_model = grid_search.best_estimator_
    y_pred = best_model.predict(X_test)
    
    accuracy = accuracy_score(y_test, y_pred)
    precision = precision_score(y_test, y_pred)
    recall = recall_score(y_test, y_pred)
    f1 = f1_score(y_test, y_pred)
    
    # Append the results to the list
    model_results.append({
        'Model': model_name,
        'Best Params': grid_search.best_params_,
        'Accuracy': accuracy,
        'Precision': precision,
        'Recall': recall,
        'F1-score': f1
    })
    
    # Generate and display confusion matrix
    conf_matrix = confusion_matrix(y_test, y_pred)
    plt.figure(figsize=(8, 5))
    sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues')
    plt.title(f'Confusion Matrix for {model_name}')
    plt.xlabel('Predicted')
    plt.ylabel('True')
    plt.show()

    # Get feature importances
    if model_name in ['Decision Tree', 'Random Forest', 'XGBoost']:
        importances = best_model.feature_importances_
        feature_importance[model_name] = importances
    elif model_name == 'Logistic Regression':
        importances = np.abs(best_model.coef_[0])
        feature_importance[model_name] = importances

# Convert the list to a DataFrame
results = pd.DataFrame(model_results)

# Display the results
print(results)
```
